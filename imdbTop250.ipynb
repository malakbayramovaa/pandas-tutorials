# Dataset haqqında ümumi məlumat

```python
# pandasi import edin ve imbd datasetini yukleyin
import pandas as pd
df = pd.read_csv("/content/imdbTop250.csv")
```

```python
# Datasetin ilk 10 sətrini ekrana çıxarın.
df.head(10)
```

```python
# Datasetin sütun adlarını göstərin.
df.columns
```

```python
# Datasetdə neçə sətrin və neçə sütunun olduğunu tapın.
df.shape
print("Row sayi: ", df.shape[0])
print("Column sayi:", df.shape[1])
```

```python
# Hər bir sütunun data tipini (dtype) göstərən kod yazın.
df.dtypes
```

```python
# Datasetdə ümumilikdə neçə null dəyər olduğunu tapın.
df.isnull().sum().sum()
```

```python
# Hər sütundakı null dəyərlərin sayını çıxarın.
df.isnull().sum()
```

```python
# Hansı sətrlərdə null dəyər varsa, həmin bütün sətrləri ekrana çıxarın.
df[df.isnull().any(axis=1)]
```

```python
# Score sütunundakı null dəyərləri median ilə doldurun.
df['Score']=df['Score'].fillna(df['Score'].median())
```

```python
df.isnull().sum()
```

```python
# Gross sutunundaki null deyerleri ortalama ile doldurun
df['Gross'] = df['Gross'].fillna(df['Gross'].mean())
```

```python
df.isnull().sum()
```

```python
# Cast3 ve Cast4 sütunlarindakı null dəyərləri əvvəlki sətirdəki dəyərlə (ffill) doldurun.
df['Cast3'] = df['Cast3'].fillna(method='ffill')
```

```python
df['Cast4'] = df['Cast4'].fillna(method='ffill')
```

```python
df.isnull().sum()
```

# Əsas statistik analiz

```python
# IMDb rating kolonunun maksimum, minimum və orta qiymətlərini tapın.
df.describe()['Rating']
```

```python
print("Maksimum rating:", df['Rating'].max())
print("Minimum rating:", df['Rating'].min())
print("Ortalama rating:", df['Rating'].mean())
```

```python
# Ən yüksək rating alan filmi tapın (title və rating).    #loc idxmax
df.loc[df['Rating'].idxmax()][['Title', 'Rating']]
```

```python
# Ən aşağı rating alan filmi tapın.
df.loc[df['Rating'].idxmin()][['Title', 'Rating']]
```

```python
# Date sütununda ən köhnə və ən yeni filmin illerini tapın.
print("en kohne film:",df['Date'].min())
print("En yeni film:", df['Date'].max())
```

# Filtrləmə və seçmə

```python
# IMDb reytinqi 8.5-dən yüksək olan filmləri seçin.
df[df['Rating']>8.5][['Title', 'Rating']]
```

```python
# 2000–2010 illəri arasında çıxmış filmləri siyahıya alın
df[df['Date'].between(2000,2010)][['Title', 'Date']]
```

```python
# Rating-i 9.0–dan yüksək olan ilk 5 filmi seçin.
df[df['Rating']>9].head(5)
```

# Qruplaşdırma və sıralama

```python
# Filmləri il üzrə qruplaşdırın və hər ilin orta IMDb reytinqini hesablayın.
df.groupby('Date')['Rating'].mean()
```

```python
# Filmləri IMDb rating-ə görə azalan sırayla düzün.
df.sort_values(by='Rating',ascending=False)
```

```python
# Rating × votes ilə “IMDB_score_power” adlı yeni sütun yaradın (yüksək reytinq + çox səs = daha güclü film).
IMDB_score_power= df['Rating']*df['Votes']
IMDB_score_power
```

```python
# Datasetdə təkrarlanan sətir (duplicate) olub-olmadığını yoxlayın və varsa silin.
df.duplicated().sum()
```

```python
# Directora görə qruplaşdırın və hər bir rejissorun neçə film çəkdiyini tapın.
# → Ən çox filmi olan ilk 5 rejissoru göstər.
qrup = df.groupby('Director')['Title'].count()

qrup.sort_values(ascending=False).head(5)
```
