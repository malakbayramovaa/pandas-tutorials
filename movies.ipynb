```python
df['release_date']=pd.to_datetime(df['release_date'], errors='coerce',format='mixed')
```

```python
df=df.drop(columns=['belongs_to_collection', 'homepage', 'original_title', 'id', 'spoken_languages'])
```

```python
df.head()
```

```python
import numpy as np
```

```python
df['budget']=pd.to_numeric(df['budget'], errors='coerce')
df['budget']=df['budget'].replace(0, np.nan)
df['budget']=df['budget'].fillna(df['budget'].median())
```

```python
df['popularity']=pd.to_numeric(df['popularity'], errors='coerce')
df['popularity']=df['popularity'].replace(0, np.nan)
df['popularity']=df['popularity'].fillna(df['popularity'].median())
```

```python
df.isnull().sum()
```

```python
numeric_cols=df.select_dtypes(include=np.number)
categoric_cols = df.select_dtypes(include='object')
```

```python
numeric_cols
```

```python
categoric_cols
```

```python
import ast
df['genres']=df['genres'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

```python
df['genres']=df['genres'].apply(lambda x: x[0]['name'] if isinstance(x, list) and len(x)>0 else np.nan)
```

```python
df
```

```python
df['production_companies']=df['production_companies'].apply(lambda x: ast.literal_eval(x) if isinstance(x,str) else x)
```

```python
df['production_companies']=df['production_companies'].apply(lambda x: x[0]['name'] if isinstance(x, list) and len(x) > 0 else np.nan)
```

```python
df
```

```python
df['production_countries']=df['production_countries'].apply(lambda x: ast.literal_eval(x) if isinstance(x,str) else x)
df['production_countries']=df['production_countries'].apply(lambda x: x[0]['iso_3166_1'] if isinstance(x,list) and len(x)>0 else np.nan)
```

```python
df
```

```python
df.isnull().sum()
```

```python
for column in numeric_cols:
  df[column]=df[column].fillna(df[column].median())
```

```python
for column in categoric_cols:
  df[column]=df[column].fillna(df[column].mode()[0])
```

```python
df.isnull().sum()
```
