# 180. Consecutive Numbers

https://leetcode.com/problems/consecutive-numbers/description/?envType=study-plan-v2&envId=top-sql-50

```python
data = [[1, 1], [2, 1], [3, 1], [4, 2], [5, 1], [6, 2], [7, 2]]
logs = pd.DataFrame(data, columns=['id', 'num']).astype({'id':'Int64', 'num':'Int64'})
```

```python
logs
```

```python
# shift(1)-->evvelki
# shift(-1)-->sonraki
```

```python
logs['previous']=logs['num'].shift(1)
logs['next']=logs['num'].shift(-1)
logs['check']=(logs['num']==logs['previous'])&(logs['num']==logs['next'])
logs
df=logs[logs['check']]['num'].unique()
pd.DataFrame({'ConsecutiveNums':df})
```

# 626. Exchange Seats

```python
data = [[1, 'Abbot'], [2, 'Doris'], [3, 'Emerson'], [4, 'Green'], [5, 'Jeames']]
seat = pd.DataFrame(data, columns=['id', 'student']).astype({'id':'Int64', 'student':'object'})
```

# 185. Department Top Three Salaries

https://leetcode.com/problems/department-top-three-salaries/?envType=study-plan-v2&envId=top-sql-50

```python
data = [[1, 'Joe', 85000, 1], [2, 'Henry', 80000, 2], [3, 'Sam', 60000, 2], [4, 'Max', 90000, 1], [5, 'Janet', 69000, 1], [6, 'Randy', 85000, 1], [7, 'Will', 70000, 1]]
employee = pd.DataFrame(data, columns=['id', 'name', 'salary', 'departmentId']).astype({'id':'Int64', 'name':'object', 'salary':'Int64', 'departmentId':'Int64'})
data = [[1, 'IT'], [2, 'Sales']]
department = pd.DataFrame(data, columns=['id', 'name']).astype({'id':'Int64', 'name':'object'})
```

```python
employee
```

```python
department
```

```python
df = employee.merge(department, left_on="departmentId", right_on="id", suffixes=("", "_dept"))

top3 = (
    df.groupby("departmentId")["salary"].apply(lambda x: sorted(x.unique(), reverse=True)[:3]).explode()
    .reset_index(name="top_salary")
)

result = df.merge(top3, left_on=["departmentId", "salary"], right_on=["departmentId", "top_salary"])

result = result[["name_dept", "name", "salary"]]
result.columns = ["Department", "Employee", "Salary"]
result
```
