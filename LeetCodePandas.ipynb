# Question 10

```python
data = [[9, 'Hercy', None, 43], [6, 'Alice', 9, 41], [4, 'Bob', 9, 36], [2, 'Winston', None, 37]]
employees = pd.DataFrame(data, columns=['employee_id', 'name', 'reports_to', 'age']).astype({'employee_id':'Int64', 'name':'object', 'reports_to':'Int64', 'age':'Int64'})
```

```python
employees
```

```python
sert1=employees.groupby('reports_to').agg(reports_count=('employee_id','size'), mean_age=('age', 'mean'))
sert1['mean_age']=sert1['mean_age'].round().astype(int)

result=sert1.merge(employees[['employee_id', 'name']],
                   left_on='reports_to',
                   right_on='employee_id',
                   how='left')

result=result[['employee_id','name','reports_count','mean_age']].sort_values('employee_id')
result
```

```python
data = [[1, 20, '2019-08-14'], [2, 50, '2019-08-14'], [1, 30, '2019-08-15'], [1, 35, '2019-08-16'], [2, 65, '2019-08-17'], [3, 20, '2019-08-18']]
products = pd.DataFrame(data, columns=['product_id', 'new_price', 'change_date']).astype({'product_id':'Int64', 'new_price':'Int64', 'change_date':'datetime64[ns]'})
```

# Question 20

https://leetcode.com/problems/count-salary-categories/?envType=study-plan-v2&envId=top-sql-50

```python
data = [[3, 108939], [2, 12747], [8, 87709], [6, 91796]]
accounts = pd.DataFrame(data, columns=['account_id', 'income']).astype({'account_id':'Int64', 'income':'Int64'})
```

```python
accounts
```

```python
def count_salary_categories(accounts: pd.DataFrame) -> pd.DataFrame:

    low=(accounts['income']<20000).sum()
    average_salary=(accounts['income'].between(20000,50000)).sum()
    high_salary=(accounts['income']>50000).sum()
    return pd.DataFrame({"category":["Low Salary","Average Salary","High Salary"],
                      "accounts_count":[low, average_salary, high_salary]})
```

https://leetcode.com/problems/list-the-products-ordered-in-a-period/?envType=study-plan-v2&envId=top-sql-50

# Question 24

https://leetcode.com/problems/movie-rating/description/?envType=study-plan-v2&envId=top-sql-50

```python
data = [[1, 'Avengers'], [2, 'Frozen 2'], [3, 'Joker']]
movies = pd.DataFrame(data, columns=['movie_id', 'title']).astype({'movie_id':'Int64', 'title':'object'})
data = [[1, 'Daniel'], [2, 'Monica'], [3, 'Maria'], [4, 'James']]
users = pd.DataFrame(data, columns=['user_id', 'name']).astype({'user_id':'Int64', 'name':'object'})
data = [[1, 1, 3, '2020-01-12'], [1, 2, 4, '2020-02-11'], [1, 3, 2, '2020-02-12'], [1, 4, 1, '2020-01-01'], [2, 1, 5, '2020-02-17'], [2, 2, 2, '2020-02-01'], [2, 3, 2, '2020-03-01'], [3, 1, 3, '2020-02-22'], [3, 2, 4, '2020-02-25']]
movie_rating = pd.DataFrame(data, columns=['movie_id', 'user_id', 'rating', 'created_at']).astype({'movie_id':'Int64', 'user_id':'Int64', 'rating':'Int64', 'created_at':'datetime64[ns]'})
```

```python
movies
```

```python
users
```

```python
movie_rating
```

```python
movie_rating["created_at"] = pd.to_datetime(movie_rating["created_at"])

u = (
    movie_rating.groupby("user_id").size().reset_index(name="cnt")
                .merge(users, on="user_id").sort_values(["cnt", "name"], ascending=[False, True])
                .iloc[0]["name"]
)

feb = movie_rating[
    (movie_rating["created_at"] >= "2020-02-01") & (movie_rating["created_at"] <= "2020-02-29")
]

m = (
    feb.groupby("movie_id")["rating"].mean().reset_index(name="avg").merge(movies, on="movie_id")
       .sort_values(["avg", "title"], ascending=[False, True]).iloc[0]["title"]
)

result = pd.DataFrame({"results": [u, m]})
result

```

# Question 25

https://leetcode.com/problems/department-top-three-salaries/?envType=study-plan-v2&envId=top-sql-50

```python
data = [[1, 'Joe', 85000, 1], [2, 'Henry', 80000, 2], [3, 'Sam', 60000, 2], [4, 'Max', 90000, 1], [5, 'Janet', 69000, 1], [6, 'Randy', 85000, 1], [7, 'Will', 70000, 1]]
employee = pd.DataFrame(data, columns=['id', 'name', 'salary', 'departmentId']).astype({'id':'Int64', 'name':'object', 'salary':'Int64', 'departmentId':'Int64'})
data = [[1, 'IT'], [2, 'Sales']]
department = pd.DataFrame(data, columns=['id', 'name']).astype({'id':'Int64', 'name':'object'})
```

```python
employee
```

```python
department
```

```python
df = employee.merge(department, left_on="departmentId", right_on="id", suffixes=("", "_dept"))

top3 = (
    df.groupby("departmentId")["salary"].apply(lambda x: sorted(x.unique(), reverse=True)[:3]).explode()
    .reset_index(name="top_salary")
)

result = df.merge(top3, left_on=["departmentId", "salary"], right_on=["departmentId", "top_salary"])

result = result[["name_dept", "name", "salary"]]
result.columns = ["Department", "Employee", "Salary"]
print(result)
```
